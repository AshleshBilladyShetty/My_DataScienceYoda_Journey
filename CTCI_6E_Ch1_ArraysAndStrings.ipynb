**1.1 Is Unique** Implement an algorithm to determine if a string has all Unique Characters. what if you cannot use any additional data structures?

**(1)** Python can have else even for "for" loop https://www.geeksforgeeks.org/using-else-conditional-statement-with-for-loop-in-python/  
**(2)** Break Continue and return in python https://www.learnpython.dev/02-introduction-to-python/110-control-statements-looping/50-exercise/

```python
%%time
def is_unique(string):
    alpha = 'abcdefghijklmnopqrstuvwxyz'
    alphalen = len(alpha)
    unique_bool = True
    for i in string:
        alpha = alpha.replace(i,'')
        if alphalen == len(alpha):
            unique_bool = False
            break
        else:
            alphalen = len(alpha)
    
    if unique_bool:
        value =  'No duplicate characters'
    else:
        value = 'Has duplicate characters'
    return print(value)
```

```python
is_unique('apple')
```
